```python
#1++3++5+.....+100= ?
sum = 0
for i in range (1, 101):
  if i % 2 == 0:
    continue
  else:
    sum = sum + i

print(sum)
```

```python
#1++3++5+.....+100= ?
sum = 0
for i in range (1, 101):
  if i % 2 != 0:

    sum = sum + i

print(sum)
```

```python
#while loop
i = 1
sum=0
while i <101:
  sum += i
  i = i + 2

print(sum)
```
